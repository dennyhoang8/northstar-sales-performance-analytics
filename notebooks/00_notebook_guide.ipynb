# Notebook Guide

This folder contains the Jupyter notebooks used throughout the NorthStar Sales Performance Analytics project.

Each notebook represents a specific phase of the data analytics workflow and builds upon the results of the previous notebook.



# Why Separate the Project into Multiple Notebooks?

Each notebook has a single responsibility and represents one phase of the data analytics workflow.

Keeping each notebook focused makes the project easier to understand, maintain, and review.

---

## 01_data_exploration.ipynb

**Purpose**

Understand the dataset before making any changes.

Use this notebook to answer questions such as:

- What columns exist?
- What are the data types?
- Are there missing values?
- Are there duplicate records?
- What does the data look like?
- What are the summary statistics?

If I ever forget the dataset structure, I should come back to this notebook.

---

## 02_data_cleaning.ipynb

**Purpose**

Clean and prepare the dataset for analysis.

Use this notebook to answer questions such as:

- Which data quality issues were found?
- How were missing values handled?
- Were duplicate records removed?
- Was the cleaned dataset successfully created?

If I ever forget what changes were made to the data, I should come back to this notebook.

---

## 03_exploratory_data_analysis.ipynb

**Purpose**

Analyze the cleaned dataset to answer business questions.

This notebook should focus on insights, not data exploration.

Since the dataset has already been explored and cleaned, there is no need to repeat functions such as:

- `df.head()`
- `df.info()`
- `df.columns`
- `df.describe()`

If I need to remember the dataset structure, I should return to **01_data_exploration.ipynb**.

---

## 04_data_visualization.ipynb

**Purpose**

Create visualizations that communicate the key findings from the exploratory data analysis.

This notebook focuses on presenting insights rather than exploring or cleaning the data.

---

# Project Workflow

```text
01_data_exploration.ipynb
            ↓
02_data_cleaning.ipynb
            ↓
03_exploratory_data_analysis.ipynb
            ↓
04_data_visualization.ipynb
```

Each notebook builds upon the previous one and has a specific role within the overall analytics workflow.

Keeping responsibilities separate results in a cleaner, more organized, and more professional project.
