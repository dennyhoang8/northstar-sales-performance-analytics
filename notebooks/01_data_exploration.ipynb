# 01 Data Exploration

This notebook explores the raw sales dataset for the NorthStar Sales Performance Analytics project.

The goal is to understand the dataset structure, review column names, check data types, identify missing values, and detect duplicate records before performing data cleaning.

````python
import pandas as pd
````

## 2. Load Dataset

The raw Excel file is stored in the `data/raw` folder.

````python
df = pd.read_excel("../data/raw/Sample - Superstore.xls")
````

## 3. Preview the Data

Display the first few rows to understand the structure of the dataset.

````python
df.head()
````

## 4. Dataset Shape

Check the number of rows and columns in the dataset.

````python
df.shape
````

## 5. Column Names

Review all column names in the dataset.

````python
df.columns
````

## 6. Dataset Information

Review data types, non-null counts, and memory usage.

````python
df.info()
````

## 7. Summary Statistics

Review summary statistics for numeric columns such as Sales, Quantity, Discount, and Profit.

````python
df.describe()
````

## 8. Missing Values

Check for missing values in each column.

````python
df.isnull().sum()
````

## 9. Duplicate Rows

Check whether the dataset contains duplicate records.

````python
df.duplicated().sum()
````

## 10. Summary Statistics for All Columns

Display summary statistics for both numeric and non-numeric columns.

````python
df.describe(include="all")
````

## 11. Unique Values

Determine the number of unique values in each column.

````python
df.nunique()
````

## 12. Random Sample

Display a random sample of records to verify that the data looks accurate and consistent throughout the dataset.

A fixed `random_state` is used to ensure the same random sample is generated each time the notebook is run, making the results reproducible and consistent.

```python
df.sample(5, random_state=42)
```

## 13. Initial Observations

### Findings

- The dataset contains **9,994** sales records and **21** columns.
- Most columns contain complete data.
- The **Postal Code** column contains **11 missing values**.
- No duplicate records were identified.
- The dataset includes a mix of numeric, categorical, and datetime fields suitable for business analysis.
- Sales and Profit exhibit a wide range of values, indicating potential outliers that will be explored in later phases.

### Conclusion

The dataset is generally well-structured and requires minimal cleaning before exploratory data analysis. The only notable issue identified during exploration is 11 missing values in the Postal Code column, while no duplicate records were detected.

## Purpose of Each Exploration Function

| Function | Question It Answers |
|----------|---------------------|
| `df.head()` | What does the dataset look like? |
| `df.shape` | How big is the dataset (rows and columns)? |
| `df.columns` | What variables (columns) are available? |
| `df.info()` | What data types do I have, and are there any missing values? |
| `df.describe()` | What do the numeric values look like (summary statistics)? |
| `df.isna().sum()` | Are there any missing values in the dataset? |
| `df.duplicated().sum()` | Are there any duplicate rows? |
| `df.nunique()` | How many unique values are in each column? |
| `df.sample()` | Does the data look reasonable throughout the dataset? |
