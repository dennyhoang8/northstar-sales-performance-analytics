# 02 Data Cleaning

This notebook cleans and prepares the raw sales dataset for analysis in the NorthStar Sales Performance Analytics project.

The goal is to address data quality issues identified during exploration, verify the cleaning results, and save a cleaned version of the dataset for future analysis.

# 1. Import Libraries

Import the required libraries for the data cleaning process.

```python
import pandas as pd
```

# 2. Load Dataset

Load the raw dataset that will be cleaned and prepared for analysis.

```python
df = pd.read_excel("../data/raw/Sample - Superstore.xls")
```

# 3. Initial Dataset Shape

Review the dataset dimensions before cleaning to establish a baseline.

```python
df.shape
```

# 4. Check Missing Values

Identify columns containing missing values that may require cleaning.

```python
df.isnull().sum().sort_values(ascending=False)
```

# 5. Review Missing Records

Inspect records with missing Postal Code values before applying a cleaning decision.

```python
df[df["Postal Code"].isnull()]
```

# 6. Cleaning Decision

The dataset contains **11 missing values** in the **Postal Code** column.

Because Postal Code is an identifier rather than a measurement and is not required for the planned exploratory analysis, SQL analysis, or Power BI dashboard, these missing values will be retained.

No imputation or row removal will be performed.

# 7. Verify Missing Values

Confirm that all missing values have been addressed.

```python
df.isnull().sum()
```

# 8. Check Duplicate Rows

Check whether the dataset contains duplicate records.

```python
df.duplicated().sum()
```

# 9. Remove Duplicate Rows

Remove duplicate records, if any exist, to ensure each row represents a unique transaction.

```python
df = df.drop_duplicates()
```

# 10. Final Dataset Shape

Review the dataset dimensions after completing the cleaning process.

```python
df.shape
```

# 11. Save Clean Dataset

Save the cleaned dataset for use in exploratory data analysis, SQL analysis, visualization, and dashboard development.

```python
df.to_csv("../data/cleaned/superstore_cleaned.csv", index=False)
```

# 12. Cleaning Summary

## Cleaning Results

- Reviewed the initial dataset shape before cleaning.
- Identified 11 missing values in the Postal Code column.
- Reviewed the records containing missing Postal Code values.
- Replaced missing Postal Code values with `"Unknown"`.
- Verified that no missing values remain.
- Checked for duplicate records.
- Removed duplicate records if any were present.
- Saved the cleaned dataset for future project phases.

## Conclusion

The dataset has been cleaned, verified, and prepared for exploratory data analysis, SQL analysis, visualization, and dashboard development.
