# 03 Exploratory Data Analysis (EDA)

This notebook explores the cleaned sales dataset to identify trends, evaluate business performance, and uncover actionable insights.

The analysis focuses on sales, profitability, customers, products, and regional performance to support data-driven business decisions.

# 1. Import Libraries

Import the required libraries for exploratory data analysis.

```python
import pandas as pd
```

# 2. Load Cleaned Dataset

Load the cleaned dataset that will be used throughout the exploratory data analysis process.

```python
df = pd.read_csv("../data/cleaned/superstore_cleaned.csv")
```

# 3. Dataset Overview

Review the dimensions of the cleaned dataset before beginning the analysis.

```python
df.shape
```

# 4. Overall Sales Performance

Calculate key sales metrics to establish a high-level understanding of the company's overall business performance.

```python
total_sales = df["Sales"].sum()

total_sales
```

### Business Insight

The total sales value represents the overall revenue generated across all recorded transactions. This metric serves as a baseline for evaluating business performance throughout the remaining analysis.

# 5. Total Profit

Calculate the total profit generated across all recorded transactions to evaluate the company's overall profitability.

```python
total_profit = df["Profit"].sum()

total_profit
```

### Business Insight

Total profit measures the company's overall earnings after accounting for costs. Comparing this value with total sales provides a high-level view of overall business performance.

# 6. Total Orders

Determine the total number of orders processed within the dataset.

```python
total_orders = df["Order ID"].nunique()

total_orders
```

### Business Insight

The total number of unique orders reflects overall customer purchasing activity and serves as a key indicator of business volume.

# 7. Average Order Value

Calculate the average sales amount generated per customer order.

```python
average_order_value = total_sales / total_orders

average_order_value
```

# 8. Sales by Region

Analyze total sales across each region to identify the strongest and weakest performing markets.

```python
sales_by_region = (
    df.groupby("Region")["Sales"]
      .sum()
      .sort_values(ascending=False)
)

sales_by_region
```

### Business Insight

Comparing total sales by region helps identify the company's strongest markets and highlights regions that may require additional sales efforts or strategic improvements.

# 9. Profit by Region

Analyze total profit across each region to evaluate overall profitability by market.

```python
profit_by_region = (
    df.groupby("Region")["Profit"]
      .sum()
      .sort_values(ascending=False)
)

profit_by_region
```

### Business Insight

Comparing profit by region reveals which markets contribute the most to overall profitability and helps identify regions where strong sales may not translate into strong financial performance.

# 10. Sales by State

Analyze total sales by state to identify the highest and lowest performing geographic markets.

```python
sales_by_state = (
    df.groupby("State")["Sales"]
      .sum()
      .sort_values(ascending=False)
)

sales_by_state
```

### Business Insight

State-level sales analysis provides a more detailed view of geographic performance, helping identify key markets that drive revenue and areas with potential for future growth.

# 11. Sales by Category

Analyze total sales across each product category to identify which categories generate the most revenue.

```python
sales_by_category = (
    df.groupby("Category")["Sales"]
      .sum()
      .sort_values(ascending=False)
)

sales_by_category
```

### Business Insight

Comparing sales by category helps identify the company's primary revenue drivers and highlights product categories with the greatest sales contribution.

# 12. Profit by Category

Analyze total profit across each product category to evaluate overall profitability.

```python
profit_by_category = (
    df.groupby("Category")["Profit"]
      .sum()
      .sort_values(ascending=False)
)

profit_by_category
```

### Business Insight

Comparing profit by category helps determine which product categories contribute the most to the company's overall profitability.

# 14. Profit by Sub-Category

Analyze total profit across each product sub-category to identify the most and least profitable product groups.

```python
profit_by_subcategory = (
    df.groupby("Sub-Category")["Profit"]
      .sum()
      .sort_values(ascending=False)
)

profit_by_subcategory
```

### Business Insight

Profit by sub-category highlights which product groups generate the strongest financial returns and which may require further evaluation to improve profitability.

# 15. Top Products by Sales

Identify the products that generate the highest total sales to understand which items contribute the most to overall revenue.

```python
top_products_sales = (
    df.groupby("Product Name")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_products_sales
```

### Business Insight

The highest-selling products represent the company's strongest revenue drivers and may indicate products with consistently high customer demand.

# 16. Top Products by Profit

Identify the products that generate the highest total profit to determine which items contribute the most to overall profitability.

```python
top_products_profit = (
    df.groupby("Product Name")["Profit"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_products_profit
```

### Business Insight

High-profit products contribute the most to the company's financial success and may represent opportunities for continued investment or promotion.

# 17. Least Profitable Products

Identify the products generating the lowest total profit to uncover items that may be reducing overall business performance.

```python
least_profitable_products = (
    df.groupby("Product Name")["Profit"]
      .sum()
      .sort_values()
      .head(10)
)

least_profitable_products
```

### Business Insight

Products with the lowest profitability may require further evaluation to determine whether pricing, discounting, or operating costs are negatively impacting performance.

# 18. Top Customers by Sales

Identify the customers who generate the highest total sales to determine the company's most valuable customers by revenue.

```python
top_customers_sales = (
    df.groupby("Customer Name")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_customers_sales
```

### Business Insight

Customers generating the highest sales represent key revenue contributors. Understanding these customers can help guide retention strategies and strengthen long-term customer relationships.

# 19. Top Customers by Profit

Identify the customers who generate the highest total profit to evaluate which customer relationships contribute the most to overall profitability.

```python
top_customers_profit = (
    df.groupby("Customer Name")["Profit"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_customers_profit
```

### Business Insight

Customers generating the highest profit provide the greatest financial value to the business and may represent opportunities for continued engagement and long-term growth.

# 20. Customer Order Frequency

Analyze the number of orders placed by each customer to identify the most active customers.

```python
customer_order_frequency = (
    df.groupby("Customer Name")["Order ID"]
      .nunique()
      .sort_values(ascending=False)
      .head(10)
)

customer_order_frequency
```

### Business Insight

Customers with the highest order frequency demonstrate strong engagement and repeat purchasing behavior, making them valuable contributors to sustained business performance.

# 21. Convert Order Date

Ensure the Order Date column is stored as a date format to support time-based analysis.

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

# 22. Sales by Year

Analyze total sales by year to identify overall business growth and long-term sales trends.

```python
sales_by_year = (
    df.groupby(df["Order Date"].dt.year)["Sales"]
      .sum()
)

sales_by_year
```

### Business Insight

Annual sales trends help evaluate business growth over time and identify years with stronger or weaker overall performance.

# 23. Sales by Month

Analyze total sales by month to identify seasonal patterns and recurring sales trends.

```python
sales_by_month = (
    df.groupby(df["Order Date"].dt.month_name())["Sales"]
      .sum()
      .reindex([
          "January", "February", "March",
          "April", "May", "June",
          "July", "August", "September",
          "October", "November", "December"
      ])
)

sales_by_month
```

### Business Insight

Monthly sales trends help identify periods of higher and lower customer demand, providing insight into seasonal business performance.

# 24. Profit by Year

Analyze total profit by year to evaluate changes in overall business profitability over time.

```python
profit_by_year = (
    df.groupby(df["Order Date"].dt.year)["Profit"]
      .sum()
)

profit_by_year
```

### Business Insight

Annual profit trends help determine whether the company's financial performance is improving over time and can reveal periods of stronger or weaker profitability.
